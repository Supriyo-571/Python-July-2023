```python
import matplotlib.pyplot as plt
import numpy as np
class subplot:
    def plot():
        x_axis_1 = np.array([4, 1, 2, 3])
        y_axis_1 = np.array([3, 8, 1, 10])

        plt.subplot(3, 3, 1)
        plt.scatter(x_axis_1,y_axis_1)

        x_axis_2 = np.array([6, 1, 2, 3])
        y_axis_2 = np.array([10, 20, 30, 40])
        mylabels = ["A", "B", "C", "D"]
        myexplode = [0.2, 0, 0, 0]


        plt.subplot(3, 3, 2)
        
        plt.pie(x_axis_2,labels = mylabels, explode = myexplode)
        plt.legend()

        x_axis_3 = np.array([0, 1, 2, 3])
        y_axis_3 = np.array([3, 8, 1, 10])

        plt.subplot(3, 3, 3)
        plt.bar(x_axis_3,y_axis_3)

        x_axis_4 = np.array([3, 1, 2, 3])
        y_axis_4 = np.array([10, 20, 30, 40])

        plt.subplot(3, 3, 4)
        plt.plot(x_axis_4,y_axis_4,marker="o",mfc="red")

        x_axis_5 = np.array([5, 1, 2, 3])
        y_axis_5 = np.array([3, 8, 1, 10])

        plt.subplot(3, 3, 5)
        plt.plot(x_axis_5,y_axis_5,marker="o",mfc="green")

        x_axis_6 = np.array([8, 1, 2, 3])
        y_axis_6 = np.array([10, 20, 30, 40])

        plt.subplot(3, 3, 6)
        plt.plot(x_axis_6,y_axis_6,marker="o",mfc="blue")

        x_axis_7 = np.array([9, 1, 2, 3])
        y_axis_7 = np.array([10, 20, 30, 40])
        
        plt.subplot(3, 3, 7)
        plt.scatter(x_axis_7,y_axis_7)

        x_axis_8 = np.array([7, 1, 2, 3])
        y_axis_8 = np.array([10, 20, 30, 40])

        plt.subplot(3, 3, 8)
        plt.scatter(x_axis_8,y_axis_8)

        x_axis_9 = np.array([6, 1, 2, 3])
        y_axis_9 = np.array([10, 20, 30, 40])

        plt.subplot(3, 3, 9)
        plt.scatter(x_axis_9,y_axis_9)

        plt.show()
subplot.plot()
```
